## **Visualization**

Univariate

```python
#histogram
import matplotlib.pyplot as plt
plt.hist(data['age'])
plt.title("HISTOGRAM FOR AGE")
```

```python
#Box plot
plt.boxplot(data['age'])
plt.title("BOX PLOT OF AGE")
```

Bivariate

```python
#Scatter plot
plt.scatter(x=data['age'],y=data['fare'])
plt.title("SCATTER PLOT FOR AGE AND FARE")
plt.xlabel("AGE")
plt.ylabel("FARE")
```

```python
#Bar plot
import seaborn as sns
sns.barplot(x=data['age'],y=data['fare'])
plt.title("BAR PLOT FOR AGE AND FARE")
plt.xlabel("AGE")
plt.ylabel("FARE")
```

Multivariate

```python
#Pair plot
sns.pairplot(data)
```

```python
#Heat map
cor=data.corr()
sns.heatmap(cor)
```

```python
# Descriptive statistics
data.describe()
```

```python
data.info()
```

```python
# Replace missing values of age with mean
data['age'].fillna(data['age'].mean(),inplace=True)
data
```

# **Outliers**

```python
#Outlier detection
plt.boxplot(data['fare'])
```

```python
per=data['age'].quantile(0.99)
per
```

```python
data=data[data['fare']<=per]
plt.boxplot(data['fare'])
```

# **Encoding Techniques**

**One-hot encoding**

```python
# Perform one-hot encoding
data_main=pd.get_dummies(data,columns=['embark_town'])
data_main
```

```python
data_main.corr()
```

```python
#heat map
plt.figure(figsize=(10,8))
sns.heatmap(data_main.corr(),annot=True)
```

```python
data_main.head()
```

**Label encoding**

```python
#Perform label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['sex']=le.fit_transform(data['sex'])
data.head()
```

```python
data['alone']=le.fit_transform(data['alone'])
data.head()
```
